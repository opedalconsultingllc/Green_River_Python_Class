```python
import pandas as pd
import numpy as np
```

## Importing a csv file into Pandas

```python
# creating a variable to hold the dataframe and read the csv results into that dataframe
# UCLA IDRE
df = pd.read_csv('https://stats.idre.ucla.edu/stat/data/hsbdemo.csv')
```

```python
df.head(5)
```

## Working with data
Each row in the data frame represents a data sample.
Each column is a variable and is named.
A common tasks is to select relevant rows from the data frame for modelling and visualisation activities.

```python
df[df['female'] == 'female'].head(5)
```

```python
df.query('read<35')
```

```python
df[(df.female == 'female') & (df.read <= 35)]
```

# Groupby

```python
df.groupby(["female"])[['read']].count()
```

```python
# plot data
import matplotlib.pyplot as plt
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','red','green','blue','orange','white','brown']
df['read'].value_counts().plot(kind='bar',title='Reading scores distribution',color=colors)
plt.show()
```

```python
#finding min and max
print('Max', df['read'].max())
print('Min', df['read'].min())
```

```python
#finding mean, mode and standard deviation
print('Mean', df['read'].mean())
print('Median', df['read'].median())
print('Mode', df['read'].mode())
print('Stdev', df['read'].std())
```

```python
# df.iloc[<row selection>],[<column selection>]
# start by selecting rows
df.iloc[1:3]
```

```python
# Note that .iloc returns a Pandas Series when one row is selected
# Select the last row in the data frame
df.iloc[-1]
```

```python
# select rows 0 to 2 and columns 1 and 2
df.iloc[0:3,1:3]
```

```python
df.iloc[:,0].head(5) # first column of data frame, id, limiting to 5 by using the function head
```

```python
df.iloc[:,-3].head(5) # third to last column of data frame, id, limiting to 5 by using the function head
```
